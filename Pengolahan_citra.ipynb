```python
from scipy import ndimage
from matplotlib import pyplot as plt
from skimage import data

image = data.coffee()
blurred_image = ndimage.gaussian_filter(image, sigma=3)
very_blurred = ndimage.gaussian_filter(image, sigma=5)

fig, ax = plt.subplots(1, 3, figsize=(16, 8,))

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original Image")
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].imshow(blurred_image, cmap='gray')
ax[1].set_title("Blurred Image (sigma=3)")
ax[1].set_xticks([])
ax[1].set_yticks([])

ax[2].imshow(very_blurred, cmap='gray')
ax[2].set_title("Very Blurred Image (sigma=5)")
ax[2].set_xticks([])
ax[2].set_yticks([])

plt.show()
```

```python
# Filter linier

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread("pinguin.jfif")
fig, ax = plt.subplots(1, 3, figsize=(16, 8))
fig.tight_layout()

# untuk meng-konvolusi kernel pada sebuah gambar kita dapat menggunakan cv.filter2D
kernel_sharpening = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
kernel_sharpening_2 = np.array([[-1, -1, -1], [-1, 10, -1], [-1, -1, -1]])

ax[0].imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
ax[0].set_title('Original Image')

sharpened = cv.filter2D(image, -1, kernel_sharpening)
ax[1].imshow(cv.cvtColor(sharpened, cv.COLOR_BGR2RGB))
ax[1].set_title('Sharpened Kernel Image')

sharpened_2 = cv.filter2D(image, -1, kernel_sharpening_2)
ax[2].imshow(cv.cvtColor(sharpened_2, cv.COLOR_BGR2RGB))
ax[2].set_title('Sharpened Kernel Image 2')

plt.show()
```
